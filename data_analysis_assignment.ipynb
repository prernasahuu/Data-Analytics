## Demographic factors influencing the User Behaviour

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Total Orders', y='Amount (USD)', data=merged_data)
plt.title('Total Order by Amount (USD)')
plt.xlabel('Total Orders')
plt.ylabel('amount (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.displot(y='Total Orders', x='Favorite Meal', data=merged_data)
plt.title('Fav Meals by Total')
plt.xlabel('Fav Meal')
plt.ylabel('Total Orders')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
sns.displot(x="Dish Name_y", y= "Duration (mins)", data=merged_data)
xtricks = plt.xticks(rotation=90)
plt.title('Dish Duration')
plt.xlabel('Dish Name')
plt.ylabel('Duration (mins)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="Dish Name_y", y= "Rating", data=merged_data)
xtricks = plt.xticks(rotation=90)
plt.title('Dish Rating')
plt.xlabel('Dish Name')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(y="Rating", x= "Age", data=merged_data)
xtricks = plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(y="Age", x= "Dish Name_y", data=merged_data)
xtricks = plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.displot(y="Age", x= "Total Orders", data=merged_data)
plt.tight_layout()
xtricks = plt.xticks(rotation=90)
plt.show()
```
